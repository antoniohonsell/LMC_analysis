# CRH Metrics Analysis

This notebook analyzes **Canonical Representational Hypothesis (CRH) Polynomial Alignment Hypothesis metrics** computed during neural network training.

## Quantities Computed

### Linear Alignment Metrics (Hb_WWT) - Backward Pass

- **ε_lin (epsilon_linear)**: Linear alignment error
  $$\varepsilon_{\text{lin}} = \frac{\|Z_c - c_{\text{lin}} \cdot H_c\|_F}{\|Z_c\|_F}$$

- **c_lin (c_linear)**: Linear alignment constant/scaling factor
  $$c_{\text{lin}} = \arg\min_c \|Z_c - c \cdot H_c\|$$

- **ρ_cos (rho_cosine)**: Cosine Similarity (normalized inner product)
  $$\rho_{\text{cos}} = \frac{\langle H, Z \rangle}{||H||_F \cdot ||Z||_F} \in [-1, 1]$$
  This is a normalized similarity measure. Values range from -1 (perfect anticorrelation) to 1 (perfect correlation).

- **Pearson α**: Pearson correlation between matrices (Herdin et al., 2005)
  $$\alpha(A, B) := \frac{1}{K} \left( \frac{1}{d^2} \sum_{ij} A_{ij}B_{ij} - \frac{1}{d^4} \sum_{ij} A_{ij} \sum_{ij} B_{ij} \right)$$
  where $\alpha \in [-1, 1]$ with $\alpha = \pm 1$ for perfect alignment

### Power-Law Alignment Metrics (Hb_WWT) - Backward Pass

- **α (alpha)**: Power-law exponent
  $$\alpha \text{ from } Z_c = c_{\text{pow}} \cdot H_c^\alpha + \varepsilon_{\text{pow}}$$

- **R²**: Coefficient of determination (goodness-of-fit)
  $$R^2 = 1 - \frac{\sum_i (y_i - \hat{y}_i)^2}{\sum_i (y_i - \bar{y})^2}$$

- **ε_pow (epsilon_power-law)**: Power-law alignment error (normalized)
  $$\varepsilon_{\text{pow}} = \|Z_c - c_{\text{pow}} \cdot H_c^\alpha\| / \|H_c\|_F$$

- **c_pow (c_power-law)**: Power-law scaling constant
  $$c_{\text{pow}} = \arg\min_c \|Z_c - c \cdot H_c^\alpha\|$$

### Datasets Analyzed
- **FASHIONMNIST**: Trained MLP on Fashion MNIST dataset
- **MNIST**: Trained MLP on standard MNIST dataset

**Note**: The CSV files only contain backward pass metrics (Hb_WWT). Forward alignment data (Hf_WWT) is not available in these files.

Each dataset is evaluated across training epochs to track how these alignment metrics evolve.

## FMNIST

```python
import pandas as pd
import numpy as np

# Load the CSV file (adjust path since notebook is in CRH subdirectory)
csv_path = "../runs_sweep_full/mnist_mlp_reg/ep800_lr_0.001_wd_0.0003_hd_2048-512_do_0_ls_0.05/FASHIONMNIST/full/seed_0/final_train/crh_pah_metrics.csv"
df = pd.read_csv(csv_path)

# Show available columns
print("Available columns:")
print(df.columns.tolist())
print(f"\nDataset shape: {df.shape}")
print()

# Extract Hb_WWT linear alignment columns - only use columns that exist
linear_cols = ['epoch', 'Hb_WWT_eps_lin', 'Hb_WWT_c_lin', 'Hb_WWT_cosine_sim_fro']
available_linear_cols = [col for col in linear_cols if col in df.columns]

if available_linear_cols:
    linear_data = df[available_linear_cols].copy()
    linear_data.columns = ['Epoch', 'ε_lin', 'c_lin', 'ρ_cos'][:len(available_linear_cols)]
    
    print("=" * 70)
    print("LINEAR ALIGNMENT (Hb_WWT)")
    print("=" * 70)
    print(linear_data.to_string(index=False))
    print()
else:
    print("No linear alignment columns found in dataset")
```

```python
# Extract Hb_WWT power-law alignment columns - only use columns that exist
power_cols = ['epoch', 'Hb_WWT_powerlaw_alpha', 'Hb_WWT_powerlaw_r2', 'Hb_WWT_pow_eps', 'Hb_WWT_pow_c']
available_power_cols = [col for col in power_cols if col in df.columns]

if available_power_cols:
    power_data = df[available_power_cols].copy()
    col_names = ['Epoch', 'α (alpha)', 'R²', 'ε_pow', 'c_pow']
    power_data.columns = col_names[:len(available_power_cols)]
    
    print("=" * 70)
    print("POWER-LAW ALIGNMENT (Hb_WWT)")
    print("=" * 70)
    print(power_data.to_string(index=False))
    print()
else:
    print("No power-law alignment columns found in dataset")
```
