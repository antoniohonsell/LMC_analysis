# CRH Metrics Analysis

This notebook analyzes **Canonical Representational Hypothesis (CRH) Plynomial Alignment Hypothesis metrics** computed during neural network training.

## Quantities Computed

### Linear Alignment Metrics (Hb_WWT)

- **ε_lin (epsilon_linear)**: Linear alignment error
  $$\varepsilon_{\text{lin}} = \|Z_c - c_{\text{lin}} \cdot H_c\|$$

- **c_lin (c_linear)**: Linear alignment constant/scaling factor
  $$c_{\text{lin}} = \arg\min_c \|Z_c - c \cdot H_c\|$$

- **ρ_fro (rho_frobenius)**: Frobenius norm distance
  $$\rho_{\text{fro}} = \|H_c - Z_c\|_F = \sqrt{\sum_{i,j} (H_c - Z_c)_{ij}^2}$$

### Power-Law Alignment Metrics (Hb_WWT)

- **α (alpha)**: Power-law exponent
  $$\alpha \text{ from } Z_c = c_{\text{pow}} \cdot H_c^\alpha + \varepsilon_{\text{pow}}$$

- **R²**: Coefficient of determination (goodness-of-fit)
  $$R^2 = 1 - \frac{\sum_i (y_i - \hat{y}_i)^2}{\sum_i (y_i - \bar{y})^2}$$

- **ε_pow (epsilon_power-law)**: Power-law alignment error
  $$\varepsilon_{\text{pow}} = \|Z_c - c_{\text{pow}} \cdot H_c^\alpha\|$$

- **c_pow (c_power-law)**: Power-law scaling constant
  $$c_{\text{pow}} = \arg\min_c \|Z_c - c \cdot H_c^\alpha\|$$

### Datasets Analyzed
- **FASHIONMNIST**: Trained MLP on Fashion MNIST dataset
- **MNIST**: Trained MLP on standard MNIST dataset

Each dataset is evaluated across training epochs to track how these alignment metrics evolve.

## FMNIST

```python
import pandas as pd
import numpy as np

# Load the CSV file (adjust path since notebook is in CRH subdirectory)
csv_path = "../runs_sweep_full/mnist_mlp_reg/ep1200_lr_0.001_wd_0.0003_hd_2048-512_do_0_ls_0.05/FASHIONMNIST/full/seed_0/final_train/crh_pah_metrics.csv"
df = pd.read_csv(csv_path)

# Extract Hb_WWT linear alignment columns
linear_cols = ['epoch', 'Hb_WWT_eps_lin', 'Hb_WWT_c_lin', 'Hb_WWT_rho_fro']
linear_data = df[linear_cols].copy()
linear_data.columns = ['Epoch', 'ε_lin', 'c_lin', 'ρ_fro']

print("=" * 70)
print("LINEAR ALIGNMENT (Hb_WWT)")
print("=" * 70)
print(linear_data.to_string(index=False))
print()
```

```python
# Extract Hb_WWT power-law alignment columns
power_cols = ['epoch', 'Hb_WWT_powerlaw_alpha', 'Hb_WWT_powerlaw_r2', 'Hb_WWT_pow_eps', 'Hb_WWT_pow_c']
power_data = df[power_cols].copy()
power_data.columns = ['Epoch', 'α (alpha)', 'R²', 'ε_pow', 'c_pow']

print("=" * 70)
print("POWER-LAW ALIGNMENT (Hb_WWT)")
print("=" * 70)
print(power_data.to_string(index=False))
print()
```
